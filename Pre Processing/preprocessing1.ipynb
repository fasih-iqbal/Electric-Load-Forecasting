**Reading each file and converting into data frame**

```python
import pandas as pd

# Initialize an empty list to store the DataFrames
demand_data = []
weather_data = []
```

```python
def drop_columns(df, columns):

    return df.drop(columns=columns, errors='ignore')
```

```python
def display_data(df):

    # Display column names and datatypes
    print("\nColumn Names:")
    print(df.columns.tolist())

    print("\nData Types:")
    print(df.dtypes)

    print("\nFirst 2 Rows:")
    print(df.head(2))
```

```python
def preprocess_json_file(path, city):
    path = path
    
    df = pd.read_json(path)
    
    
    display_data(df)

    df['timestamp'] = pd.to_datetime(df['time'], unit='s')

    df['timestamp'] = pd.to_datetime(df['timestamp'])

    columns_to_drop = [
        'summary',
        'icon',
        'apparentTemperature',
        'dewPoint',
        'windGust',
        'windBearing',
        'cloudCover',
        'uvIndex',
        'visibility',
        'precipType',
        'ozone',
        'precipAccumulation',
        'time'
    ]

    df = drop_columns(df, columns_to_drop)

    df['city'] = city 

    return df
```

# Merging the demands and weather data

```python
# Combine all demand and weather data into single DataFrames
all_demand = pd.concat(demand_data, ignore_index=True)
all_weather = pd.concat(weather_data, ignore_index=True)

# Merge the datasets on timestamp and city
data = pd.merge(all_demand, all_weather, on=['timestamp', 'city'], how='inner')
```
